## 課題3
- 日本の各年の輸出入の合計のデータを作成してみましょう

```
with sqlite3.connect('/content/y_1997.db') as conn:
  sql = 'select exp_imp, Year, sum(Value) as Value from year_from_1997 group by exp_imp, Year'
  jp1 = pd.read_sql(sql, conn)
jp1
```

## 課題4
- PlotlyExpressのサイトを見ながら、何か他のグラフを作成してみてください
    - https://plotly.com/python/plotly-express/

```
boueki_dict = {1: "輸出", 2: "輸入"}
jp1["exp_imp"] = jp1['exp_imp'].map(boueki_dict)
jp2 = jp1.pivot(index="Year", columns="exp_imp", values="Value")
```

```
jp2['balance'] = jp2["輸出"] - jp2["輸入"]
```

```
jp2
```

```
px.bar(jp2, x=jp2.index, y='balance', title="Japanese_trade_balance")
```

## 課題5
- 便利に貿易収支を見られるDashアプリケーションを作ってみよう

```
app = JupyterDash(__name__)

app.layout = html.Div([
                       
                       html.H1(id='title'),

                       dcc.Dropdown(id='my_drop',
                                    options=[{"label": col, 'value': col} for col in jp2.columns],
                                    value='輸出'
                                    ),
                       dcc.Graph(id='my_graph')
])

@app.callback([Output('title', 'children'), Output('my_graph', 'figure')], [Input('my_drop', 'value')])
def update_graph(selected_value):
  return f'日本の貿易収支データ（{selected_value}）', px.bar(jp2, x=jp2.index, y=selected_value)

app.run_server(mode='inline')
```
